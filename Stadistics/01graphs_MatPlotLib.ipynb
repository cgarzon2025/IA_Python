## **Gráficas con MatPlotLib**

### **Creación de Gráficas de Líneas (plot)**

```python
import matplotlib.pyplot as plt

axis_x = ['a', 'b', 'c', 'd', 'e']
dates1 = [10, 20, 30, 40, 50]
dates2 = [20, 30, 40, 50, 60]

plt.plot(axis_x, dates2)
plt.plot(axis_x, dates1)
plt.show()

plt.plot(axis_x, dates2)
plt.show()
```

## **Generar Datos a Grafica**
### Examen de Certificación para Estudiantes de 5 niveles

```python
import numpy as np

ages = ['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']

level1 = np.random.rand(10)*100
level2 = np.random.rand(10)*200 + 100
level3 = np.random.rand(10)*300 + 200  
level4 = np.random.rand(10)*400 + 300
level5 = np.random.rand(10)*500 + 400
```

### Personalización de Gráficas

```python
# Introducir series de datos con color, marcador, etiqueta y estilo de linea

plt.plot(ages, level1, label='Nivel 1', color='red', marker='o', linestyle='-')
plt.plot(ages, level2, label='Nivel 2', color='blue', marker='x', linestyle='--')
plt.plot(ages, level3, label='Nivel 3', color='green', marker='s', linestyle=':')
plt.plot(ages, level4, label='Nivel 4', color='black', marker='d', linestyle='-.')
plt.plot(ages, level5, label='Nivel 5', color='purple', marker='*', linestyle='')

# Activar la leyenda
plt.legend()

# Título de la gráfica
plt.title('Examen de Certificación para Estudiantes de 5 niveles')

# Etiqueta de los ejes
plt.xlabel('Años')
plt.ylabel('Niveles')

# Personalizar Eje de Puntaje de 0 a 1000 en incrementos de 50
plt.yticks(np.arange(0, 1001, 50))

# Activar la cuadricula
plt.grid()

# Activar marcas menores
plt.minorticks_on()
plt.show()
```

## **Gráficas de barras verticales (bar)**

```python
axis_x = np.arange(0, 10)

plt.bar(axis_x, level5, width=1/5)
plt.bar(axis_x + 0.2, level4, width=1/5)
plt.bar(axis_x + 0.4, level3, width=1/5)
plt.bar(axis_x + 0.6, level2, width=1/5)
plt.bar(axis_x + 0.8, level1, width=1/5)

plt.show()
```

## **Gráfica de barras horizontales (barh)**

```python
plt.barh(axis_x + 0.8, level5, height=1/5)
plt.barh(axis_x + 0.6, level4, height=1/5)
plt.barh(axis_x + 0.4, level3, height=1/5)
plt.barh(axis_x + 0.2, level2, height=1/5)
plt.barh(axis_x, level1, height=1/5)

plt.show()
```

## **Gráfica de barras verticales apiladas**

```python
plt.bar(axis_x, level5)
plt.bar(axis_x, level4, bottom=level5)
plt.bar(axis_x, level3, bottom=level5 + level4)
plt.bar(axis_x, level2, bottom=level5 + level4 + level3)
plt.bar(axis_x, level1, bottom=level5 + level4 + level3 + level2)

plt.show()
```

## **Gráfica de Dispersión (scatter)**

```python
plt.scatter(axis_x, level1, marker='o', color='red')
plt.scatter(axis_x, level2, marker='x', color='blue')
plt.scatter(axis_x, level3, marker='s', color='green')
plt.scatter(axis_x, level4, marker='d', color='black')
plt.scatter(axis_x, level5, marker='*', color='purple')

plt.show()
```

## **Gráfica de Pie**

```python
plt.pie(level1, labels= ['Colombia', 'Perú', 'Chile', 'Argentina', 'Venezuela', 'Ecuador', 'Bolivia', 'Paraguay', 'Uruguay', 'Guyana'], autopct='%1.1f%%')

plt.show()
```

## **Combinar Gráficas**

```python
plt.bar(axis_x, level1, width=1/5)
plt.bar(axis_x + 0.2, level2, width=1/5)

plt.plot(ages, level3)

plt.scatter(axis_x, level4)
plt.scatter(axis_x, level5)

plt.show()
```
